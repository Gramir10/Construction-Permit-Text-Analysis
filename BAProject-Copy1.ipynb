```python
#libraries
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.figure_factory as ff
```

```python
#data
df_HP = pd.read_csv("Housing_Permits.csv", dtype={'PermitID': object})
df_HP.info()
```

```python
df_HP.head()
```

```python
#Removing Rows that have Null Values
df_HP = df_HP[df_HP["PermitID"].notnull()]
df_HP = df_HP[df_HP["CaseNum"].notnull()]
df_HP = df_HP[df_HP["Block"].notnull()]
df_HP = df_HP[df_HP["Lot"].notnull()]
df_HP = df_HP[df_HP["PropertyAddress"].notnull()]
df_HP = df_HP[df_HP["PermitNum"].notnull()]
df_HP = df_HP[df_HP["DateIssue"].notnull()]
df_HP = df_HP[df_HP["PermitDescription"].notnull()]
df_HP = df_HP[df_HP["Cost_Est"].notnull()]
df_HP = df_HP[df_HP["DateExpire"].notnull()]
df_HP = df_HP[df_HP["Prop_Use"].notnull()]
df_HP = df_HP[df_HP["Existing_Use"].notnull()]
df_HP = df_HP[df_HP["Neighborhood"].notnull()]
df_HP = df_HP[df_HP["PoliceDistrict"].notnull()]
df_HP = df_HP[df_HP["CouncilDistrict"].notnull()]
df_HP = df_HP[df_HP["Location"].notnull()]
df_HP = df_HP[df_HP["2010 Census Neighborhoods"].notnull()]
df_HP = df_HP[df_HP["2010 Census Wards Precincts"].notnull()]
df_HP = df_HP[df_HP["Zip Codes"].notnull()]
```
